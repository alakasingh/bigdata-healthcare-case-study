```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('healthcare_dataset.csv')

# Clean column names (remove spaces)
df.columns = df.columns.str.strip()

print(df.columns)  # check exact column names
```

.

ðŸ”¹ Step 1: Gender Distribution

Weâ€™ll see how many Male vs Female patients there are.

```python


plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df, palette='cool')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.show()
```

ðŸ”¹ Step 2: Age Distribution

Weâ€™ll check the age range and most common age groups.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True, color='purple')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

ðŸ”¹ Step 3: Average Age by Gender

This helps us see if male or female patients tend to be older.

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Gender', y='Age', data=df, palette='Set2', estimator='mean')
plt.title('Average Age by Gender')
plt.show()
```
